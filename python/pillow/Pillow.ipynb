# Pillow Library


Pillow is a fork from the Python Imaging Library (PIL) project that is no longer maintained.  
It provides methods to open, edit and save images in Python.

## Imports

Only the `PIL` module needs to be imported to use Pillow.  
Jupyter is able to render a PIL image natively when giving the image as the last instruction of a code block.

```python
from PIL import Image 
```

## Open an image

```python
with Image.open('./bulldog.jpg') as img:
    img.load()

img.size       # width and height :  (810, 225)
img.mode       # RGB (screen display) / RGBA / CMYK (Cyan Magenta Yellow Black, for ink printing) / L (gray scale)
img.format     # PNG / JPEG

img
```

## Crop an image

```python
cropped_img = img.crop((30, 10, 140, 130))    # width_start, height_start, width_end, height_end
cropped_img
```

## Resize an image

```python
small_img = img.resize((img.width, img.height // 2))
small_img
```

## Transform an image

```python
# basic transformations :
# - flip :          FLIP_LEFT_RIGHT, FLIP_TOP_BOTTOM
# - rotation :      ROTATE_90, ROTATE_180, ROTATE_270
# - transposition : TRANSPOSE (columns and rows are flipped using top-left pixel unchanged)
# - transverse :    TRANSVERSE  (columns and rows are flipped using bottom-left pixel unchanged)
transposed_img = img.transpose(Image.TRANSPOSE)
transposed_img
```

```python
# Rotation with a custom angle
rotated_img = img.rotate(20)  # rotate and keep original dimension (fill background with black)
rotated_img = img.rotate(20, expand=True)  # rotate and increase size to cover the entire image
rotated_img
```

## Change an image mode or format

```python
gray_img = img.convert('L')    # grayscale mode
gray_img
```

```python
# An image is constituted of 1 or more band(s) per pixel
img.getbands()        # ('R', 'G', 'B')
gray_img.getbands()   # ('L',)
```

```python
# To save the image in a different format, just specify the extension when saving the file
gray_img.save('gray_img.png')
```

## Access Image Pixels

```python
img.getpixel((100, 120))     # return RGB values
```

```python
gray_img.getpixel((100, 120))     # return gray value
```

```python
for x in range(20, 100):
    cropped_img.putpixel((x, 50), (255, 0, 0))    # draw a red line

cropped_img
```

## Display in Matplotlib

PIL images can be opened in Matplotlib if we need to use plot tools on them..

```python
import matplotlib.pyplot as plt

plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```
